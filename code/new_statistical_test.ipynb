```python
import pandas as pd
from scipy.stats import pearsonr

# Create the DataFrame
df = pd.DataFrame({
    'grade': [6.5, 6.5, 6.5, 6.5, 6.5,
              5.5, 7.5, 7.5, 8.5, 9.5,
              8.5, 8.5],
    'ai_knowledge': [1, 2, 3, 4, 5,
                     3, 2, 3, 3, 3,
                     4, 5]
})

# Perform Pearson correlation
corr, p_value = pearsonr(df['ai_knowledge'], df['grade'])

print("Pearson's correlation coefficient:", corr)
print("p-value:", p_value)
```

```python
import numpy as np
import scipy.stats as stats

# Correlation r and sample size
r = 0.21
n = 51

# Fisher Z-transformation to calculate the confidence interval
z = np.arctanh(r)
se = 1 / np.sqrt(n - 3)
z_critical = stats.norm.ppf(0.975)  # 95% confidence interval

# Calculate confidence interval in Z-space and transform back to r
z_lower = z - z_critical * se
z_upper = z + z_critical * se
r_lower = np.tanh(z_lower)
r_upper = np.tanh(z_upper)

print(f"95% CI for r: ({r_lower:.2f}, {r_upper:.2f})")
```
